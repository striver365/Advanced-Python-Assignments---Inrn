Question 1. 

You are given two strings s and t. String t is generated by randomly
shuffling string s and then adding one more letter at a random position.
Return the letter that was added to t.

Examples

find_the_difference(&quot;abcd&quot;, &quot;abcde&quot;) ➞ &quot;e&quot;

find_the_difference(&quot;&quot;, &quot;y&quot;) ➞ &quot;y&quot;

find_the_difference(&quot;ae&quot;, &quot;aea&quot;) ➞ &quot;a&quot;

```python
def find_the_difference(str_a, str_b):

#     str_a = "ae"
#     str_b = "aea"
    s1 = s2 = ""
    c = 0
    if len(str_a) > len(str_b):
        s1 = str_a
        s2 = str_b
    else:
        s1 = str_b
        s2 = str_a

    for i in (s1):
        if s1.count(i) != s2.count(i):
            print(i)
            break
```

```python
find_the_difference("", "y")
```

```python
find_the_difference("ae", "aea")
```

Question 2.

Given a function that accepts unlimited arguments, check and count how
many data types are in those arguments. Finally return the total in a list.
List order is:

[int, str, bool, list, tuple, dictionary]

Examples

count_datatypes(1, 45, &quot;Hi&quot;, False) ➞ [2, 1, 1, 0, 0, 0]

count_datatypes([10, 20], (&quot;t&quot;, &quot;Ok&quot;), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0]

count_datatypes(&quot;Hello&quot;, &quot;Bye&quot;, True, True, False, {&quot;1&quot;: &quot;One&quot;, &quot;2&quot;: &quot;Two&quot;}, [1,
3], {&quot;Brayan&quot;: 18}, 25, 23) ➞ [2, 2, 3, 1, 0, 2]

count_datatypes(4, 21, (&quot;ES&quot;, &quot;EN&quot;), (&quot;a&quot;, &quot;b&quot;), False, [1, 2, 3], [4, 5, 6]) ➞ [2,
0, 1, 2, 2, 0]

```python
def count_datatypes(*data):
    int_ = str_ = bool_ = lst_ = tupl_  = dict_ = dtypes = 0
    lst1 = []

    for i in data:
        lst1.append(i)
    for i in lst1:
        if type(i) == int:
            int_ += 1
        elif type(i) == str:
            str_ += 1
        elif type(i) == bool:
            bool_ += 1
        elif type(i) == list:
            lst_+= 1
        elif type(i) == tuple:
            tupl_ += 1
        elif type(i) == dict:
            dict_ += 1
    dtypes = int_, str_, bool_, lst_, tupl_, dict_
    print(list(dtypes))
```

```python
count_datatypes(1, 45, "Hi", False)
```

```python
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

```python
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

```python
def count_dtypes(*some_data):
    one_lst = []

    for i in some_data:
        one_lst.append(i)
        print(type(i))

    c_int =  list(map(type, one_lst)).count(int)
    c_str =  list(map(type, one_lst)).count(str)
    c_bool = list(map(type, one_lst)).count(bool)
    c_lst = list(map(type, one_lst)).count(list)
    c_tupl = list(map(type, one_lst)).count(tuple)
    c_dict = list(map(type, one_lst)).count(dict)

    coll_ = c_int, c_str, c_bool, c_lst, c_tupl, c_dict
    return list(coll_)
```

```python
count_dtypes(1, 45, "Hi", False)
```

```python
count_dtypes([10, 20], ("t", "Ok"), 2, 3, 1)
```

```python
count_dtypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

```python
count_dtypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

A Fibonacci string is a precedence of the Fibonacci series. It works with
any two characters of the English alphabet (as opposed to the numbers 0 and
1 in the Fibonacci series) 

as the initial items and concatenates them together
as it progresses in a similar fashion as the Fibonacci series.

Examples

fib_str(3, [&quot;j&quot;, &quot;h&quot;]) ➞ &quot;j, h, hj&quot;

fib_str(5, [&quot;e&quot;, &quot;a&quot;]) ➞ &quot;e, a, ae, aea, aeaae&quot;

fib_str(6, [&quot;n&quot;, &quot;k&quot;]) ➞ &quot;n, k, kn, knk, knkkn, knkknknk&quot;

```python
def fib_str(num_, str_lst):
    
    first = str_lst[0]
    second = str_lst[1]
    count = 0
    temp = ""
    
    while count < num_:
        print(first)
        temp = second + first
        first = second
        second = temp
        count += 1
```

Question 4. 

Given an integer between 0 and 26, make a variable (self.answer). That
variable would be assigned to a string in this format:

&quot;nines:your answer, threes:your answer, ones:your answer&quot;

You need to find out how many ones, threes, and nines it would at least take
for the number of each to add up to the given integer when multiplied by one,
three or nine (depends).

Examples

ones_threes_nines(10) ➞ &quot;nines:1, threes:0, ones:1&quot;

ones_threes_nines(15) ➞ &quot;nines:1, threes:2, ones:0&quot;

ones_threes_nines(22) ➞ &quot;nines:2, threes:1, ones:1&quot;

```python
def ones_threes_nines(num):
    val = 0
    nines = []
    threes = []
    ones = []

    if num > 0 and num <= 26:
        val = num // 9
        nines.append(val)
        num = num - (val*9)
        val = num // 3
        threes.append(val)
        num = num - (val*3)
        ones.append(num) 
    else:
        print("Number should be more than 0 ")
    print("Nines: ", nines, "Threes: ", threes, "Ones : ", ones)
```

```python
ones_threes_nines(10)
```

```python
ones_threes_nines(15)
```

```python
ones_threes_nines(22)
```

```python
def one_3_9(num):
    nines = num // 9
    threes = (num%9) // 3
    ones = num % 3
    
    print("Nines:", nines)
    print("Threes:", threes)
    print("Ones:", ones)
```

the value of the sequence at a given index is dependent on the sum of the
last two values.

Therefore it can be important to
think about the implications of running a function recursively. 

Depending on

want to take a different approach.

Fibonacci sequence&#39;s value at index n.

0 + 1 = 1, 1 + 1 = 2, 1 + 2 = 3, 2 + 3 = 5, 3 + 5 = 8

fib(2) ➞ 1
